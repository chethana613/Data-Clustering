```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
```

```python
print(iris.data.shape)
print(iris.DESCR)
```

# Scatter plot every pair of features

```python
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(15,10))

attributes = ["sepal_length", "sepal_width", "petal_length", "petal_width"]


flattened_axes = axs.flatten()
axis_index = 0

for i, attr1 in enumerate(attributes):
  for j, attr2 in enumerate(attributes[i+1:]):
    ax = flattened_axes[axis_index]
    ax.scatter(X[:,i],X[:,j+i+1])
    ax.set_xlabel(attr1)
    ax.set_ylabel(attr2)
    axis_index += 1
```

```python
from sklearn.cluster import KMeans

num_clusters = 3
kmeans = KMeans(n_clusters=num_clusters).fit(iris.data)
cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_
labels
```

```python
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(15,10))

attributes = ["sepal_length", "sepal_width", "petal_length", "petal_width"]


flattened_axes = axs.flatten()
axis_index = 0

for i, attr1 in enumerate(attributes):
  for j, attr2 in enumerate(attributes[i+1:]):
    ax = flattened_axes[axis_index]
    ax.scatter(X[:,i],X[:,j+i+1], c=labels)
    ax.set_xlabel(attr1)
    ax.set_ylabel(attr2)
    axis_index += 1
```
